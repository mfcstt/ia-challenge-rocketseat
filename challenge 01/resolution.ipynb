## Exercício do módulo
Com base nas aulas do módulo de Estatística, utilize os métodos da biblioteca Pandas para tratar o dicionário abaixo e:
- Trazer a média das vendas
- Criar um gráfico de barras vertical mostrando o mês de referência e o valor
- Criar um gráfico de linhas mostrando o mês de referência e o valor.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
df_faturamento
```

### Média

```python
mean_sales = round(df_faturamento['valor'].mean(),2)
mean_sales
```

### Gráfico de barras vertical mostrando o mês de referência e o valor

```python
df_faturamento.plot.bar(x='data_ref', y='valor', color='purple')
```

### Linhas mostrando o mês de referência e o valor.

```python
df_faturamento.plot.line(x='data_ref', y='valor', color='purple')
```
